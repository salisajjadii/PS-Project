```python
import numpy as np
import matplotlib.pyplot as plt

locations = []
cLoc = 0
for i in range(0, 1000):
    y = np.random.uniform(0, 1)
    if y >= 0.6:
        cLoc += 1
    if y <= 0.4:
        cLoc -= 1
    locations.append(cLoc)

plt.scatter(np.linspace(0, 1000, 1000), locations, color='gray', edgecolors='black', label='(B, A) = (0.4, 0.6)')
plt.xlabel('Steps       (movings ~ Uniform(0, 1))')
plt.ylabel('Current Position')
plt.legend()
plt.show()
```
