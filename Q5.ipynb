```python
import numpy as np
import matplotlib.pyplot as plt 

p = 0.45
q = 1 - p
n = np.linspace(0, 20, 21)
Pn = []

for i in range(0, 21):
    Pn.append((((q/p) ** i) - 1) / (((q/p) ** 20) - 1))

plt.scatter(n, Pn, color='black', label='P=0.45')
plt.xlabel('n Values')
plt.ylabel('Pn Values')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

p = 0.50
q = 1 - p
n = np.linspace(0, 20, 21)
Pn = []

for i in range(0, 21):
    Pn.append((i * ((q/p) ** (i-1)) - 1) / (20 * ((q/p) ** 19) - 1))

plt.scatter(n, Pn, color='black', label='P=0.5')
plt.xlabel('n Values')
plt.ylabel('Pn Values')
plt.legend()
plt.show()
```
