```python
import numpy as np
import matplotlib.pyplot as plt

currentLocation = 100
newLocations = []
for i in range(0,10000):
    randomNumbers = np.random.normal(0, 1, 10)
    newLocations.append(currentLocation+sum(randomNumbers))

count = sum(1 for x in newLocations if x > 105)
plt.scatter(np.linspace(0,10000,10000), newLocations, color='gray', edgecolors='black', label='n=10')
plt.plot(range(10000), [105]*10000, color='red')
plt.title(f'The number of times the final position is passed (105) is {count}')
plt.ylabel('Final position')
plt.legend()
plt.show()
```
