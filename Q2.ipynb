```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tickets = []
x = np.arange(1, 10001)
for i in range (0, 10000):
    new = 0
    for j in range (0, 80):
        new += random.randint(1, 10)      
    tickets.append(new)
data = pd.DataFrame({'Row_value':tickets})
count = (data['Row_value'] > 300).sum()
print(f"More than (300) = {count}")
plt.scatter(x, tickets, color='gray')
print()
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(70, 150)
totalTickets = []
for k in range(70, 150):
    tickets = 0
    for i in range (0, 10000):
        new = 0
        for j in range (0, k):
            new += random.randint(1, 10)      
        if (new >= 300):
            tickets += 1
    totalTickets.append(tickets)
for i in range(len(totalTickets)):
    totalTickets[i] /= 10000

plt.scatter(x, totalTickets, color='gray')
```
