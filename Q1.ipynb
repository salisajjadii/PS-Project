```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb 

n = 10
p = 0.1
x = np.arange(0, 16, 1)
y = []

for k in range (0 , 16):
    if n >= k-1:
        y.append(comb(n, k) * ((1-p)**(k)) * (p ** (n-k)))
    else :
        y.append(0)


plt.scatter(x, y, color='black', label='p=0.1 & n=10')
plt.xlabel('k value')
plt.ylabel('fx')
plt.legend()
plt.show()
```
