```python
import random
import matplotlib.pyplot as plt

cLoc = [0, 0]
mList = [[1, 0], [-1, 0], [0, 1], [0, -1]]
x = []
y = []
dx = []
dy = []
n = 10

x.append(cLoc[0])
y.append(cLoc[1])

for i in range(0, n):
    cLoc = [x + y for x, y in zip(cLoc, random.choice(mList))]
    x.append(cLoc[0])
    y.append(cLoc[1])
    if i > 0:
        dx.append(x[i] - x[i-1])
        dy.append(y[i] - y[i-1])
        plt.quiver(x[i-1], y[i-1], dx[i-1], dy[i-1], angles='xy', scale_units='xy', scale=1, color='red')
plt.scatter(x, y, color='black', label=f'For n={n}')
plt.title('The movement and direction diagram, starts from [0, 0]')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

cLoc = [0, 0]
x = []
y = []
dx = []
dy = []
n = 10
mean = 0
std = 1

x.append(cLoc[0])
y.append(cLoc[1])

for i in range(0, n):
    cLoc = [x + y for x, y in zip(cLoc, [np.random.normal(mean, std), np.random.normal(mean, std)])]
    x.append(cLoc[0])
    y.append(cLoc[1])
    if i > 0:
        dx.append(x[i] - x[i-1])
        dy.append(y[i] - y[i-1])
        plt.quiver(x[i-1], y[i-1], dx[i-1], dy[i-1], angles='xy', scale_units='xy', scale=1, color='red')
plt.scatter(x, y, color='black', label=f'For n={n}')
plt.title('The movement and direction diagram, starts from [0, 0]')
plt.xlabel(f'X axis             steps ~ [N({mean}, ({std})**2), N({mean}, ({std})**2)]')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```
